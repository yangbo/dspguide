让我们生成一个 Hamming 窗和 Blackman 窗

```python
import numpy as np
import matplotlib.pyplot as plt
np.arange(10)
```

```python
def hamming(M):
    """Hamming 窗，长度是 M"""
    i = np.arange(M)
    w = 0.54 - 0.46*np.cos(2*np.pi*i/M)
    return w

def blackman(M):
    """Blackman 窗，长度是 M"""
    i = np.arange(M)
    w = 0.42 - 0.5*np.cos(2*np.pi*i/M)+0.08*np.cos(4*np.pi*i/M)
    return w
```

```python
M=20
plt.plot(hamming(M), '-*')
plt.plot(blackman(M), '.-')
plt.legend(('hamming','blackman'))
plt.title("Hamming,Blackman Window, Length=%d" % M)
```

可以看到 Blackman 窗y值从0开始，而 hamming 窗 y 值不从 0 开始。
下面来看看最重要的 sinc() 函数。

```python
def sinc(fc, M):
    """sinc 函数，这里 fc 是截止频率，是 (0,1) 间的数，是 fc*Fs/2 才是最终的频率"""
    N=M-1
    # 因为 arange 的结束下标是闭区间，不包含关系，所以要多加 1
    n1=np.arange(-N/2,0)
    n2=np.arange(1,N/2+1)
    w1=np.sin(2*np.pi*fc*n1)/(np.pi*n1)
    w2=np.sin(2*np.pi*fc*n2)/(np.pi*n2)
    return np.hstack((w1,2*fc,w2))
```

```python
M=101
y=sinc(0.07, M)
N=M-1
plt.plot(np.arange(-N/2,N/2+1), y, '.-')
```

可以看到 sinc() 函数的频率的确是个(0,1)间的小数，并且当i=0时，h[n]=2*fc 看上去比较合适，但其他资料说应该是 1 。还不知道什么原因，待查。

```python
# 另外，意外得到一个特殊函数!
def sinc(fc, M):
    """sinc 函数，这里 fc 是截止频率，是 (0,1) 间的数，是 fc*Fs/2 才是最终的频率"""
    N=M-1
    # 因为 arange 的结束下标是闭区间，不包含关系，所以要多加 1
    n1=np.arange(-N/2,0)
    n2=np.arange(1,N/2+1)
    w1=np.sin(2*np.pi*fc*n1)/np.pi*n1
    w2=np.sin(2*np.pi*fc*n2)/np.pi*n2
    return np.hstack((w1,np.sin(fc),w2))
M=101
y=sinc(0.1, M)
N=M-1
plt.plot(np.arange(-N/2,N/2+1), y, '.-')
```
